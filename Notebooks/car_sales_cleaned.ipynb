```python
# Import Dependancies
import pandas as pd
```

```python
# Load and read csv file
data = ('/Users/chac/Desktop/Data/car_price/Data/car_prices.csv')
df = pd.read_csv(data)
```

```python
# Display dataset
df
```

```python
# Remove timezone information if it's not needed
df['saledate'] = df['saledate'].str.replace(r' GMT.*', '', regex=True)

# Convert the 'saledate' column to string to avoid mixed types
df['saledate'] = df['saledate'].astype(str)

# Attempt to convert 'saledate' to datetime with error coercion
df['saledate'] = pd.to_datetime(df['saledate'], errors='coerce')

# Check for rows where saledate could not be parsed
invalid_dates = df[df['saledate'].isna()]

# Display invalid date entries
print("\nInvalid date entries:")
print(invalid_dates)

# If you want to drop these invalid entries
df_cleaned = df.dropna(subset=['saledate']).copy()  # Create a cleaned copy

# Extract the year from 'saledate'
df_cleaned['purchase_year'] = df_cleaned['saledate'].dt.year

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df_cleaned[['saledate', 'purchase_year']])
```

```python
# Display full dataset
df_cleaned
```

```python
# Drop NAN/Null Values
df_cleaned = df_cleaned.dropna()
```

```python
# Display updated dataframe
df_cleaned
```

```python
# Export dataframe as csv to be used in visualizations
df_cleaned.to_csv('/Users/chac/Desktop/Data/car_price/Outputs/car_sales_cleaned.csv', index=False)
```
